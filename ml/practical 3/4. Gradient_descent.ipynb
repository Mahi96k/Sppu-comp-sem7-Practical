```python
def derivative(x):
    return 2*(x+3)
```

```python
def gradient_descent(alpha, start,max_iter):
    x_list = list()
    x = start;
    x_list.append(x)
    for i in range(max_iter):
        gradient = derivative(x);
        x = x - (alpha*gradient);
        x_list.append(x);
        return x_list
```

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

def gradient_descent(expr, alpha, start, max_iter):
    x = sym.symbols('x')
    grad = sym.Derivative(expr, x).doit()
    x_val = start
    x_list = [x_val]

    for _ in range(max_iter):
        gradient = grad.subs(x, x_val)
        x_val = x_val - alpha * gradient
        x_list.append(float(x_val))

    return x_list

# Parameters
alpha = 0.1
start = 2
max_iter = 30

# Expression
x = sym.symbols('x')
expr = (x + 3)**2

# Run gradient descent
x_list = gradient_descent(expr, alpha, start, max_iter)

# Convert symbolic expr to numeric function
objective = sym.lambdify(x, expr, 'numpy')

# Plot
x_cordinate = np.linspace(-15, 15, 100)
plt.plot(x_cordinate, objective(x_cordinate), label="Objective Function")
plt.plot(x_list, objective(np.array(x_list)), 'ro-', label="Descent Path")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent Example")
plt.show()
```
